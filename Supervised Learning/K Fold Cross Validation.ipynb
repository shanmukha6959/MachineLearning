Evaluating Model Performance
- For example iris flower dataset problem we can classify using **SVM, logistic regression, Random Forest, Decision tree**. Which model out of these is the best. Cross Validation technique will help you decide which model is the best.

- **Option 1:** Use all available data for training and test on same dataset. -- (not that effective)

- **Option 2:** Split available dataset into **training** and **test** sets. --- (training dataset can be completely of different topic compared to test dataset so can also be sometimes not effective)

- **Option 3:** K fold cross validation

#### K fold cross validation
- Firstly divide the dataset into K folds and then train (K-1) folds and test with the left out 1 fold.
- Repeat the same step K times by taking each fold once as test and rest as training dataset.
- During each step note down the score.
- Finally take the average of all the scores

#### We can also do this by defining a function:

```python
def get_score(model, X_train, X_test, Y_train, Y_test):
    model.fit(X_train, Y_train)
    return model.score(X_test, Y_test)
```

```python
get_score(LogisticRegression(), X_train, X_test, Y_train, Y_test)
```

```python
get_score(SVC(), X_train, X_test, Y_train, Y_test)
```

```python
get_score(RandomForestClassifier(), X_train, X_test, Y_train, Y_test)
```

#### Exercise

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris = load_iris()
```

#### Logistic Regression

```python
lr_scores = cross_val_score(LogisticRegression(), iris.data, iris.target)
```

```python
lr_scores
```

```python
np.average(lr_scores)
```

#### SVM

```python
svm_scores = cross_val_score(SVC(), iris.data, iris.target)
```

```python
svm_scores
```

```python
np.average(svm_scores)
```

#### Random Forest 

```python
rf_scores = cross_val_score(RandomForestClassifier(), iris.data, iris.target)
```

```python
rf_scores  
```

```python
np.average(rf_scores)
```

#### Decision tree

```python
dt_scores = cross_val_score(DecisionTreeClassifier(), iris.data, iris.target)
```

```python
dt_scores
```

```python
np.average(dt_scores)
```

```python
print(f"lr: {np.average(lr_scores)}")
print(f"svm: {np.average(svm_scores)}")
print(f"rf: {np.average(rf_scores)}")
print(f"dt: {np.average(dt_scores)}")
```
