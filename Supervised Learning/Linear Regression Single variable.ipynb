```python
plt.title("canada per capitia income")
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(candy['year'], candy['per capita income (US$)'], color = "green", marker = "*")
```

```python
regressor = linear_model.LinearRegression()
regressor_train = regressor.fit(candy[['year']], candy['per capita income (US$)'])
```
