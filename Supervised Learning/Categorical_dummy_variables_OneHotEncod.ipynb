# Categorical Variables
- Nominal: Categories those dont have any numeric ordering.
    
    Ex: 
        male, Female
        green, red, blue
        town1, town2, town3
- Ordinal: Categories those have some sort of numeric ordering.
    
    Ex: 
        High, Medium, Low
        Satisfied, Neutral, Dissatisfied
        Graduate, Masters, PHD

# OneHotEncoding

### Using Pandas we can create dummies

```python
dummies = pd.get_dummies(hp.town)
```

```python
dummies
```

### we can use concat to merge two tables

```python
merged = pd.concat([hp,dummies], axis = "columns")
```

```python
merged
```

- we can drop the town column as we dont use it because we already have dummy variable columns.
- we need to drop one of the dummy columns because when a variable can be derived from rest of the variables it is called **multicolinear** and it create the problem of dummy variable trap.
- for example if we have 5 dummy variable columns we can drop 1 dummy variable column.

#### we can drop those columns using drop condition

```python
hp_final = merged.drop(["town", "west windsor"], axis = "columns")
```

```python
hp_final
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

- We can get the independent variables by dropping the dependent variables

```python
X = hp_final.drop("price",axis = "columns")
```

```python
X
```

- We can assign a different variable for dependent variables

```python
Y = hp_final.price
```

```python
Y
```

```python
reg.fit(X, Y)
```

```python
reg.predict([[2000,0,1]])
```

```python
reg.predict([[3400,0,0]])
```

- We can find how well the predicted model is performing by calculating the score and if the score is 1 the model is performing exactly as the original one. but if it is lessthan 1 the is not completely perfect as original one. 

```python
reg.score(X,Y)
```

### OneHotEncoding

```python
hp
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
hple = hp
# le.fit_transform(hple.town) O:array([0, 0, 0, 0, 0, 2, 2, 2, 2, 1, 1, 1, 1])
hple.town = le.fit_transform(hple.town)
```

```python
hple
```

```python
X = hple[["town", "area"]].values
```

```python
X
```

```python
Y = hple.price
```

```python
Y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder= 'passthrough')
```

```python
X = ct.fit_transform(X)
```

```python
X
```

```python
X = X[:,1:]
```

```python
X
```

```python
reg.fit(X,Y)
```

```python
reg.predict([[0,1,3400]])
```

```python
reg.predict([[1,0,2800]])
```
